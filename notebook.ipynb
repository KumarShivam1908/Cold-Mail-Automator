```python
import json

# Load JSON
with open("C:\\Users\\KShivam\\OneDrive - Stryker\\Desktop\\ok\\outputs\\yc-founders.json", "r", encoding="utf-8") as f:
    companies = json.load(f)

seen = set()
duplicates = []
unique_companies = []

for company in companies:
    company_name = company.get("name", "").strip()

    if company_name not in seen:
        seen.add(company_name)
        unique_companies.append(company)
    else:
        duplicates.append(company_name)

print(f"Original companies : {len(companies)}")
print(f"Unique companies   : {len(unique_companies)}")
print(f"Duplicate records  : {len(duplicates)}")

print("\nDuplicate company names:")
for name in sorted(set(duplicates)):
    print("-", name)

# Save unique companies
with open("yc-founders-unique-companies.json", "w", encoding="utf-8") as f:
    json.dump(unique_companies, f, indent=2, ensure_ascii=False)

print("\nSaved to yc-founders-unique-companies.json")
```

```python
import json

with open("yc-founders-unique-companies.json", "r", encoding="utf-8") as f:
    companies = json.load(f)

# Order of batches within a year
season_order = {
    "Winter": 3,
    "Summer": 2,
    "Spring": 1,
    "Fall": 0
}

def sort_key(company):
    batch = company.get("batch", "").strip()

    try:
        season, year = batch.split()
        return (int(year), season_order.get(season, -1))
    except:
        return (0, -1)

# Newest -> Oldest
companies = sorted(companies, key=sort_key, reverse=True)

with open("yc-founders-sorted.json", "w", encoding="utf-8") as f:
    json.dump(companies, f, indent=2, ensure_ascii=False)

print("Saved to yc-founders-sorted.json")
```
